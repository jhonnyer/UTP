# Concatenacion y union de un Dataframe

Panda ofrece dos funciones que son concat y merge

* La funcion concat nos permite concatenar un dataframe a partir de un eje determinado
* La funcion merge permite realizar uniones (joins) entre dataframe como se realizan en una base de datos
* Metodo append ofrece una funcionalidad similar a concat pero reducidad, es decir solo se maneja en el eje 0 y unicamente trabaja con la union 'outer'

```python
import numpy as np
df1=pd.DataFrame(np.arange(9).reshape([3,3]),
                index=['a','b','c'],
                columns=['A','B','C'])
df1
```

```python
import numpy as np
df2=pd.DataFrame(np.arange(12).reshape([4,3]),
                index=['a','b','c','d'],
                columns=['B','C','D'])
df2
```

```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [5, 6, 7, 8]})
print(df1)
print()
print(df2)
print()
df1.merge(df2, left_on='lkey', right_on='rkey')
```

```python
pd.concat([df1,df2],sort=True)
```

```python
pd.concat([df1,df2], axis=1,sort=True)
```

```python
#Va a realizar la union de ambos dataframe teniendo en cuenta una etiqueta comun de ambos dataframe
pd.concat([df1,df2],join='inner')
```

```python
#Va a realizar la union de ambos dataframe teniendo en cuenta una etiqueta comun de ambos dataframe
pd.concat([df1,df2],axis=1,join='inner')
```

```python
#Calcular la ganancia de notas mediante un algoritmo que multiplique el numero de 
#notas por un numero entre 1 y 5 y crear una nueva columna con ese valor
import random
def Notas(nota):
    ganancia=nota*random.randint(1,5)
    return ganancia

#Funcion apply, por cada valor en la columna notas le voy a aplicar la funcion que
#se creo llamada Notas y la almacena en la columna que se crea nueva 
#llamada ganancia
df=df.fillna({'notas':0}) #A los valores nulos que encuentre en el dataframe les voy a colocar por defecto el valor de 0
df['ganancia']=df['notas'].apply(Notas)

df.head(10)
```

```python
#Funcion que crea una nueva columna diferencia, pero que pasa a la funcion los valores
#de las filas notas y ganancia
def diferencia(fila):
    resultado=fila['notas']-fila['ganancia']
    return resultado

#axis=1 siggnifica que la funcion diferencia se va a aplicar a cada fila del dataframe
#Fila contiene el dataframe df, pero se aplica a las columnsa notas y ganancia
df['diferencia']=df.apply(diferencia, axis=1)
df.head()
```

# Escribir un archivo de texto linea a linea

```python
# Escribir archivos de texto en python

data=['Linea 1','Linea 2','Linea 3','Linea 4','Linea 5']
```

```python
fic=open('text_1.txt','w')

for line in data:
    fic.write(line)
    fic.write("\n")
fic.close()
```

```python
# Una forma de escribir el archivo línea a línea sin que sea necesario incluir el salto de línea es con la
# función print. Para esto es necesario incluir la opción "file" con la conexión al archivo
fic=open('text_2.txt','w')

for line in data:
    print(line,file=fic)
fic.close()
```

```python
fic = open("text_3.txt", "w")
fic.writelines("%s\n" % s for s in data)
fic.close()
```

```python
fic = open('text_1.txt', "r")
lines = fic.readlines()
fic.close()
lines

# En esta ocasión lines es un vector en el que cada elemento es una línea del archivo. Alternativamente,
# en lugar del método readlines se puede usar la función list para leer los datos
```

```python
fic = open('text_1.txt', "r")
lines = list(fic)
fic.close()
lines
```

# Eliminar los saltos de línea en el archivo importado:
Los tres métodos que se han visto para leer los archivos importan el salto de línea. Por lo que puede ser
necesario eliminarlo antes de trabajar con los datos. Esto se puede conseguir de forma sencilla con el
método rstrip de las cadenas de texto de Python. Lo que se puede hacer iterando sobre el vector.

```python
lines1 = [s.rstrip('\n') for s in lines]
lines1
```
