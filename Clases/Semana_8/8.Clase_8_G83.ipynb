```python
import matplotlib.pyplot as plt

x1=[2,5,6,14]
y1=[11,22,33,44]

x2=[2,5,6,15]
y2=[4,12,18,45]

#2 graficas lineales en una misma rejilla o cuadricula
plt.plot(x1,y1,color='blue',linewidth=3,label='Linea 1')
plt.plot(x2,y2,color='green',linewidth=3,label='Linea 2')

#Vamos a colocar el titulo de la grafica
plt.title('Dos gráficas lineales juntos')
#Etiquetas o titulos a los ejes
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
#Vamos a mostrar las etiquetas de cada una de las graficas
plt.legend()
#MEtodo grid me permite crear la rejilla o la cuadricula de la grafica
plt.grid()
#Mostramos la grafica en pantalla
plt.show()
```

# Grafica de barras

```python
plt.bar(x1,y1,color='blue',linewidth=3,label='Barra 1')
plt.bar(x2,y2,color='green',linewidth=3,label='Barra 2')

#Vamos a colocar el titulo de la grafica
plt.title('Dos gráficas lineales juntos')
#Etiquetas o titulos a los ejes
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
#Vamos a mostrar las etiquetas de cada una de las graficas
plt.legend()
#MEtodo grid me permite crear la rejilla o la cuadricula de la grafica
plt.grid()
#Mostramos la grafica en pantalla
plt.show()
```

# Vamos a generar un histograma

```python
#Histograma
Datos=[20,22,21,20,23,25,28,40,22,23,22,15,16,18,18,19,21,22,24,4,12,17,17,22,30]
#Rangobin es el tipo de agrupacion que yo quiero obtener de los datos a  mostrar en el histograma
Rangobin=[0,10,20,20,30,40]
# Rangobin=2
# rwidth=0.8 es la separacion entre cada barra
plt.hist(Datos,Rangobin,histtype='bar',rwidth=0.8, color='lightgreen')

#En el eje horizontal se colocan los valores de los limites de la clase, en este caso Rangobin
#En el eje vertical se colocan los valores de la frecuencia absoluta, que me indica el rango de valores
# que se encuentra los valores de datos segun los limites de la clase
#Entonces por ejemplo entre 0 y 10 se encuentra 1 valor
#Entre 10 y 20 se encuentran 8 valores
#Entre 20 y 30 se encuentra 14 valores 
#Entre 30 y 40 se encuentra 2 valores

plt.title('Histograma')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
#plt.grid() #La grip no es necesaria
plt.show()
```

# Graficas con dataframe

```python
import pandas as pd
df=pd.DataFrame({
    'Nombre':['John','Maria','Pedro','Jenifer','Bob','Lisa','Jose'],
    'Edad':[23,78,22,19,45,33,20],
    'Ciudad':['Bogota','Medellin','Bogota','Medellin','Bogota','Armenia','Armenia'],
    'Num_niños':[2,0,0,3,2,1,5],
    'Num_mascotas':[5,1,0,5,2,2,5]
})

df
```

```python
import matplotlib.pyplot as plt
#Voy a realizar una grafica lineal
df['Num_niños'].plot(kind='line')
plt.xlabel('Numero de niños')
plt.ylabel('Ciudad')
plt.show()
```

```python
# Generar un grafico de dispersion, este permite encontrar patrones en una base de datos, una grafica
#de dispersion esta compuesta por puntos entre la intersecion del eje x y el eje y
# scatter este parametro hace referencia a una grafica de dispersion
df.plot(kind='scatter',x='Num_niños',y='Num_mascotas',color='red')
plt.show()
```

```python
# Generar a partir de un dataframe la grafica de barras
print(df)
df.plot(kind='bar',x='Nombre',y='Edad',color='blue')
plt.show()
```

```python
# Dos graficas lineales a partir del atributo kind='line'
ax=plt.gca() #Permite obtener los ejes actuales de mi grafica, se los paso a ambas graficas lineales que voy
#a construir para que se visualice en un solo grafico
print(ax)
df.plot(kind='line',x='Nombre',y='Num_niños',ax=ax)
df.plot(kind='line',x='Nombre',y='Num_mascotas',color='red',ax=ax)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.DataFrame({
    'name':[
        'john','lisa','peter','carl','linda','betty'
    ],
    'date_of_birth':['01/21/1988','03/10/1977','07/25/1999','01/22/1977','09/30/1968','09/15/1970']
})

df
```

```python
df['data_of_birth']=pd.to_datetime(df['date_of_birth'],infer_datetime_format=True)
print(df)
```

```python
# Vamos a verificar en un mes cuantas personas cumplen años y lo vamos a graficar
df['data_of_birth'].map(lambda d:d.month)
```

```python
# Vamos a verificar en un mes cuantas personas cumplen años y lo vamos a graficar
df['data_of_birth'].map(lambda d:d.month).plot(kind='hist')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import date2num
import datetime

x=[
    datetime.datetime(2011,1,4,0,0),
    datetime.datetime(2011,1,5,0,0),
    datetime.datetime(2011,1,6,0,0)
]

print(x)

x=date2num(x) #Verificar funcion date2num
print()
print(x)
y=[4,9,2]
z=[1,2,3]
k=[11,12,13]
ax=plt.subplot(111)
print(ax)
ax.bar(x-0.2,y,width=0.2,color='b',align='center')
ax.bar(x,z,width=0.2,color='g',align='center')
ax.bar(x+0.2,k,width=0.2,color='r',align='center')
ax.xaxis_date()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
#importar datos para crear dataframe
liquido_url='https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
liquido_column_headers=['Alcohol','Acido Malico','Cenizas','Alcalinidad de las cenizas',
                        'Magnesio','Fenoles totales,Flavanoides','Fenoles no flavonoides','Proantocianinas',
                       'Intensidad de color','Matiz','0D280/0D315 de sustancia diluida','Prolina']
liquido_df=pd.read_csv(liquido_url,names=liquido_column_headers)
liquido_df.head(3)
```

```python
#Figura
fig,ax1=plt.subplots()
print(fig)
print(ax1)
fig.set_size_inches(13,10) #Tamaño de la grafica


#Etiquetas de mi figura
ax1.set_xlabel('Alcohol')
ax1.set_ylabel('Intensidad de color')
ax1.set_title('Relación entre la intensidad de color y la cantidad de alcohol en el liquido')

#Puedo modificar el color en la grafica a partir de los datos de la intensidad de color
c=liquido_df['Intensidad de color']
print(c)

#Generar una grafica de dispersion
plt.scatter(liquido_df['Alcohol'],liquido_df['Intensidad de color'],c=c,cmap='RdPu')
cbar=plt.colorbar()
cbar.set_label('Intensidad de color')
```
