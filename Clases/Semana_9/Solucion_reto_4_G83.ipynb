# Calculo Diccionario 1, filtro porcentaje de contagiados por sexo


```python
data.describe() #DAtos estadisticos del dataframe de contagios
```

```python
#Funcion groupby que permite agrupar columnas con un valor en especifico de otra columna o un valor estadistico
PorcenSexo=data.groupby('Sexo').agg({'Sexo':'count'}) #Aqui la funcion count va contar cuantas personas son de sexo femenino y cuantas de sexo masculino
PorcenSexo
```

```python
# Vamos a convertir la letra f minuscula en mayuscula, y la letra m en M mayuscula 
data['Sexo']=data['Sexo'].replace('f','F')
data['Sexo']=data['Sexo'].replace('m','M')
#Funcion groupby que permite agrupar columnas con un valor en especifico de otra columna o un valor estadistico
PorcenSexo=data.groupby('Sexo').agg({'Sexo':'count'}) #Aqui la funcion count va contar cuantas personas son de sexo femenino y cuantas de sexo masculino
PorcenSexo
```

```python
# 9960 -> 100
# 4483 -> x

# x= (4483*100)/9960

PorcenSexo['Porcentaje']=(PorcenSexo['Sexo']*100)/len(data['Sexo'])
PorcenSexo=PorcenSexo.round(4) #Voy a redondear a 4 decimales el valor de porcentaje
PorcenSexo
```

```python
#Para convertir un dataframe a un diccionario, voy a utilizar la funcion to_dict() de pandas
dict1=PorcenSexo.to_dict()
print(dict1)
print()
dict1=dict1['Porcentaje']
print(dict1)
```

```python
#Filtrar los datos que sean importados
porcen=data[data['Tipo']=='Importado']
porcen.head(2)
#AGrupacion de valores por pais de procedencia y el numero de contagiados por paises de procedencia
#Count permite contar el numero de contagiados, en este caso por pais de procedencia
porcen=porcen.groupby('País de procedencia').agg({'País de procedencia':'count'}) 
porcen.head(2)

#Regla de 3
# 813 son el total de contagiados del extranjero
# 813 ->100
# 10 -> x
# x=(10*100)/813
porcen['Porcentaje']=(porcen['País de procedencia']*100)/len(data[data['Tipo']=='Importado'].index)
porcen=porcen.round(4)
porcen.head(2)
```

```python
#Para convertir un dataframe a un diccionario, voy a utilizar la funcion to_dict() de pandas
dict3=porcen.to_dict()
print(dict3)
print()
dict3=dict3['Porcentaje']
print(dict3)
```
