```python
#Condicion anidadada ejemplo
x=-10
if 0<x:
    if x<10:
        print("X es un número positivo de un solo digito")
    else:
        print("X es un número positivo de dos digitos")
else:
    print("X es un número negativo")
```

```python
#Condicion encadenada ejemplo
x=9
if 0<x and x<10:
    print("X es un número positivo de un solo digito")
elif 0<x and x>=10:
    print("X es un número positivo de dos digitos")
else:
    print("X es un número negativo")
```

```python
#Recorrer una cadena de caracteres
fruta='banana'
longitud=len(fruta)
cont=0
print("\n")
for i in range(longitud):
    print(fruta[i])
    if fruta[i]=='a':
        cont+=1
print("La letra a esta repetida {} en la frase {}".format(cont,fruta))
```
