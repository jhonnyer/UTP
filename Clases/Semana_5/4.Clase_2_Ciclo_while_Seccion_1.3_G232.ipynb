# Factorial de un numero
Factorial 6

1*2*3*4*5*6


```python
#Encontrar el factorial de un numero
def factorial(n:int)->int:
    resultado=1
    numero_actual=2
    while numero_actual<=n:
        resultado=resultado*numero_actual
        numero_actual+=1
    return resultado

factorial(6)
```

```python
#Encontrar el factorial de un numero
def factorial(n:int)->int:
    resultado=1
    numero_actual=2
    iteraciones=1
    while not numero_actual>n:
        resultado=resultado*numero_actual
        numero_actual+=1
        print("Iteraciones: ",iteraciones)
        print("Resultado temporal: {}. Numero actual: {} ".format(resultado, numero_actual))
        iteraciones+=1
    
    print("El factorial del numero {} es: {}".format(n,resultado))
    return resultado
factorial(6)
```

```python
#Funcion que me encuentra el promedio de unas notas ingresadas por el teclado
promedio, total, contar= 0.0,0,0
print("Introzca una nota de un estudiante(-1 para salir): ")
grado=int(input())
while grado!=-1:
    total=total+grado
    contar=contar+1
    print("Introzca una nota de un estudiante(-1 para salir): ")
    grado=int(input())
promedio=total/contar
print("Promedio de las notas es: "+str(promedio))    
```

```python
#Funcion que me encuentra el promedio de unas notas ingresadas por el teclado, ciclo while con un else
promedio, total, contar= 0.0,0,0
mensaje="Introzca una nota de un estudiante(-1 para salir): "
grado=int(input(mensaje))
while grado!=-1:
    total=total+grado
    contar=contar+1
    grado=int(input(mensaje))
else:
    promedio=total/contar
    print("Promedio de las notas es: "+str(promedio))    
```

```python
#Ciclo while con una sentencia break: romper un bucle si una condicion se cumple
c=0
while c<=5:
    c+=1
    if c==3:
        print("El bucle ha sido roto cuando c tiene un valor de: ",c)
        break
    print("c tiene un valor de: ",c)
else:
    print("Fin del ciclo")
```

```python
#Ciclo while con una sentencia continue, romper el ciclo while y va a pasar a la siguiente iteracion
c=0
while c<=5:
    c+=1
    if c==3:
        print("El bucle ha sido roto cuando c tiene un valor de: ",c)
        continue
    print("c tiene un valor de: ",c)
else:
    print("Fin del ciclo")
```

```python
#Ejempo menú interactivo con un ciclo while
print("Bienvenido al menú interactivo")
while True:
    print("""¿Qué funcion quieres realizar? Elige una opción
    1). Saludar
    2). Sumar dos números
    3). Salir
    """)
    opcion=input()
    if opcion=='1':
        print("Bienvenido al curso de programación")
    elif opcion=='2': #SINO SI
        n1=float(input("Ingrese el primero número a sumar"))
        n2=float(input("Ingrese el segundo número a sumar"))
        sumar=n1+n2
        print("El resultado de la suma {}+{} es: {}".format(n1,n2,sumar))
    elif opcion=='3':
        print("Hasta luego")
        break
    else:
        print("Comando desconocido, vuelve a introducir una opción \n")
```

```python
### Calculadora
n1=float(input("Introduzca el primer número"))
n2=float(input("Introduzca el segundo número"))

while True:
    print("""Elige una opcion que quieras realizar
    1). Sumar dos números
    2). Restar los dos números ingresado
    3). Multiplicar los dos números
    4). Salir de la aplicacion
    """)
    opcion=input()
    if opcion=='1':
        suma=n1+n2
        print("La suma de {}+{} es: {}".format(n1,n2,suma))
    elif opcion=='2':
        resta=n1-n2
        print("La resta de {}-{} es: {}".format(n1,n2,resta))
    elif opcion=='3':
        multiplicacion=n1*n2
        print("La multiplicación de {}*{} es: {}".format(n1,n2,multiplicacion))
    elif opcion=='4':
        print("Fin de la aplicación")
        break
    else:
        print("Comando desconocido, vuelve a ingresar una opción")
```
