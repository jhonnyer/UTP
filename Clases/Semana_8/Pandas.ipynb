# Transformacion de datos

```python
#Calcular la ganancia de retweets mediante un algoritmo que multiplique el numero de 
#retweets por un numero entre 3 y 5 y crear una nueva columna con ese valor
import random
def calcularGanancias(retweets):
    ganancia=retweets*random.randint(3,5)
    return ganancia

#Funcion apply, por cada valor en la columna retweets le voy a aplicar la funcion que
#se creo llamada calcularGanancias y la almacena en la columna que se crea nueva 
#llamada ganancia
df=df.fillna({'retweets':0})
df['ganancia']=df['retweets'].apply(calcularGanancias)

df.head(10)
```

```python
#Funcion que crea una nueva columna popularidad, pero que pasa a la funcion los valores
#de las filas followees y followers
def popularidad(fila):
    resultado=fila['followees']/fila['followers']
    return resultado

#axis=1 siggnifica que la funcion popularidad se va a aplicar a cada fila del dataframe
#Fila contiene el dataframe df, pero se aplica a las columnsa followees y followers
df['popularidad']=df.apply(popularidad, axis=1)
df.head()
```

# Agrupar datos

```python
#Va a agrupar datos por ciudad, agrupando datos en una sola fila para cada ciudad
#En este caso se agrupo a partir de los valores medios 
df.groupby('country').mean()
```

```python
#Agrupacion de datos, mediante metodo agg

#Voy a indicar en cada columna que funcion realizar, para ello en lugra de media (mean)
#Voy a indicar que sume (sum) los valores de la columna followers, la media (mean) de 
#las menciones y el maximo de retweets

df.groupby('country').agg({
    'followers':'sum', #La suma de los datos en la columna followers
    'mentions':'mean', #La media de las menciones
    'retweets':'max' #El valor maximo de retweets
})
```

```python
#Agrupacion de datos, mediante metodo agg

#Voy a indicar en cada columna que funcion realizar, para ello en lugra de media (mean)
#Voy a indicar que sume (sum) los valores de la columna followers, la media (mean) de 
#las menciones y el maximo de retweets

#La agrupacion voy a almacenar en una variable grouped para filtrar
grouped=df.groupby('country').agg({
    'followers':'sum', #La suma de los datos en la columna followers
    'mentions':'mean', #La media de las menciones
    'retweets':'max' #El valor maximo de retweets
})

print(grouped)
print()
#Filtrar por condicion donde followers sea mayor a 5000
grouped[grouped['followers']>5000]
```
