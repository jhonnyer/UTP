# Histogramas con la funcion plt.hist()

Un histograma simboliza la distribución de un conjunto de datos. Sirven para obtener una "primera vista" general, o panorama, de la distribución de la población, o de la muestra, respecto a una característica, cuantitativa y continua.

```python
#Histograma
Datos=[20,22,21,20,23,25,28,40,22,23,22,15,16,18,18,19,21,22,24,4,12,17,17,22,30]
Rangobin=[0,10,20,20,30,40]
plt.hist(Datos,Rangobin,histtype='bar',rwidth=0.8,color='lightgreen')

#En el eje horizontal se colocan los valores de los limites de la clase, en este caso Rangobin
#En el eje vertical se colocan los valores de la frecuencia absoluta, que me indica el rango de valores
# que se encuentra los valores de datos segun los limites de la clase
#Entonces por ejemplo entre 0 y 10 se encuentra 1 valor
#Entre 10 y 20 se encuentran 8 valores
#Entre 20 y 30 se encuentra 14 valores 
#Entre 30 y 40 se encuentra 2 valores

plt.title('Histograma')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
#plt.grid() #La grip no es necesaria
plt.show()
```

# PIE PLOT PYTHON

```python
# Pie es unaa grafica circular
valores=[20,40,60,80]

# autopct='%1.1f%% coloquelo en forma porcentual; 1 para los enteros y 1 flotante
# explode=(0.1,0,0,0) permite la rebanada de un pedazo de la grafica o la torta, en este caso el referente
# a prekinder
# shadow=True hace referencia a la region sombreada debajo de una linea, shadow = True mostrará una 
#sombra debajo de cada etiqueta en el gráfico circular.
# startangle posicion desde donde va a arrancar la torta, se tiene en cuenta el eje horizontal del corte
# gira el inicio del gráfico circular en grados determinados en sentido antihorario desde el eje x.

plt.pie(valores,labels=['Prekinder','Kinder','Primaria','Secundaria'],colors=['red','purple','blue','orange'],
        startangle=90,shadow=True,explode=(0.1,0,0,0),autopct='%1.1f%%')

plt.title('Gráfico circular')
plt.show()
```

# Groupby-Series

El método pandas.DataFrame.groupby tiene una funcionalidad semejante a la vista para series, con los
condicionantes propios de los dataframes: es necesario indicar el eje que contiene el criterio por el que
se va a realizar la agrupación.

```python
# Serie cuyo indice es la ciudad y el valor es la cantidad de nombres unicos
# Devuelve DataFrame con recuentos de elementos únicos en cada posición.

print(type(df.groupby('Ciudad')['Nombre'].nunique()))
```

```python
#Busca coincidencias en el DataFrame
print(df)
df.groupby('Ciudad').nunique()
```

```python
#Muestra cuantas veces se repite una ciudad en el dataframe
df.groupby('Ciudad')['Nombre'].nunique()
```

```python
#Muestra cuantas veces se repite una ciudad en el dataframe
df.groupby('Ciudad')['Nombre'].nunique().plot(kind='bar')
plt.show()
```

```python
df.groupby('Ciudad')['Nombre'].nunique().plot(kind='bar')
plt.show()
```

```python
#Crea grafica por edades, size muestra grafico con edades  por ciudad de un solo color
# unstack muestra por cada ciudad un grafico con la edad de diferentes colores
df.groupby(['Ciudad','Edad']).size().unstack().plot(kind='bar')
plt.show()


# La función de Pandas dataframe.groupby () es una de las funciones más útiles en la biblioteca, 
# divide los datos en grupos basados ​​en columnas / condiciones y luego aplica algunas operaciones, 
# por ejemplo. size () que cuenta el número de entradas / filas en cada grupo. El groupby () también
# se puede aplicar en series.
```

```python
# stacked=True  muestra las graficas por edades y cada ciudad en una sola barra
#Eje derecho el numero de personas con esa edad
print(df)
df.groupby(['Ciudad','Edad']).size().unstack().plot(kind='bar',stacked=True)
plt.show()
```

# Dataframe con campo tipo fecha

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.DataFrame({
    'name':[
        'john','lisa','peter','carl','linda','betty'
    ],
    'date_of_birth':['01/21/1988','03/10/1977','07/25/1999','01/22/1977','09/30/1968','09/15/1970']
})

df
```

```python
print(df['date_of_birth'].map(lambda d:d.month))
```

```python
#Se cambio la sepracion al formato de fecha en la columna de cumpleaños
df['date_of_birth']=pd.to_datetime(df['date_of_birth'],infer_datetime_format=True)
print(df)
plt.clf() #Investigar clf

# lambda arguments : expression

# Serie. mapa ( arg , na_action = Ninguno )[fuente]
# Asignar valores de Serie de acuerdo con la correspondencia de entrada.

# (lambda d:d.month) recuperamos el mes de la fecha de cumpleaños

df['date_of_birth'].map(lambda d:d.month).plot(kind='hist')
print()
print(df)
plt.show()
```
