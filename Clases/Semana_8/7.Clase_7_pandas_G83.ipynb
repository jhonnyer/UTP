# Concatenacion y union de un Dataframe

* La funcion concat nos permite concatenar un dataframe a partir de un eje determinado, asimismo podemos utilizar un tipo de union especifico que puede ser inner o outer.
* MEtodo que es el append, este metodo ofrece una funcionalidad similiar, pero limitado al metodo concat, ya que solamente actua sobre el 0 y con el tipo de union outer

```python
import numpy as np
df1=pd.DataFrame(np.arange(9).reshape([3,3]),
                index=['a','b','c'],
                columns=['A','B','C'])
df1
```

```python
import numpy as np
df2=pd.DataFrame(np.arange(12).reshape([4,3]),
                index=['a','b','c','d'],
                columns=['B','C','D'])
df2
```

```python
pd.concat([df1,df2]) #Utiliza por defecto el metodo de union outer, que quiere decir que no tiene en cuenta
#etiquetas comunes para la union de dos dataframe
```

```python
pd.concat([df1,df2],axis=1)
```

```python
# HEmos dicho que el metodo concat permite concatenar a partir del eje 0 (filas) o del eje 1 (columnas).
#Pero tambien permite escoger el tipo de union que puede ser outer o inner.
pd.concat([df1,df2],axis=0,join='inner')
```

```python
# HEmos dicho que el metodo concat permite concatenar a partir del eje 0 (filas) o del eje 1 (columnas).
#Pero tambien permite escoger el tipo de union que puede ser outer o inner.
pd.concat([df1,df2],axis=1,join='inner')
```

```python
import random
df['notas']=random.randint(1,5)
df
```

```python
#Calcular la ganancia de notas mediante un algoritmo que multiplique el numero de 
#notas por un numero entre 1 y 5 y crear una nueva columna con ese valor
import random
def Notas(nota):
    ganancia=nota*random.randint(1,5)
    return ganancia

#Funcion apply, por cada valor en la columna notas le voy a aplicar la funcion que
#se creo llamada Notas y la almacena en la columna que se crea nueva 
#llamada ganancia
df=df.fillna({'notas':1}) # Metodo fillna hace que si en la columna notas encuentra un valor NaN, lo cambia por un valor de 0
df['ganancia']=df['notas'].apply(Notas)

df.head(10)
```

```python
#Funcion que crea una nueva columna diferencia, pero que pasa a la funcion los valores
#de las filas notas y ganancia
def diferencia(fila):
    resultado=fila['notas']-fila['ganancia']
    return resultado

#axis=1 siggnifica que la funcion diferencia se va a aplicar a cada fila del dataframe
#Fila contiene el dataframe df, pero se aplica a las columnsa notas y ganancia
df['diferencia']=df.apply(diferencia, axis=1)
df.head()
```

# Escribir en un archivo de texto linea a linea

```python
#Escribir archivos de texto, creamos una lista con los contenidos de cada linea

data=['Linea 1','Linea 2','Linea 3','Linea 4','Linea 5']
```

```python
fic=open('text_1.txt','w')
#El metodo write no aade el salto de linea, por lo cual
#es necesario aadirlo ('\n')
for line in data:
    fic.write(line)
    fic.write("\n")
fic.close()
```

```python
# Una forma de escribir el archivo lnea a lnea sin que sea necesario incluir el salto de lnea es con la
# funcin print. Para esto es necesario incluir la opcin "file" con la conexin al archivo
fic=open('text_2.txt','w')

for line in data:
    print(line,file=fic)
fic.close()
```

```python
fic = open("text_3.txt", "w")
fic.writelines("%s\n" % s for s in data)
fic.close()
```

## Leer el archivo de una vez:
El procedimiento para leer los archivos de texto ms sencillo es hacerlo de una vez con el mtodo
readlines. Una vez abierto el archivo solamente se ha de llamar a este mtodo para obtener el contenido.

```python
fic = open('text_1.txt', "r")
lines = fic.readlines()
fic.close()
lines

# En esta ocasin lines es un vector en el que cada elemento es una lnea del archivo. Alternativamente,
# en lugar del mtodo readlines se puede usar la funcin list para leer los datos
```

```python
fic = open('text_1.txt', "r")
lines = list(fic)
fic.close()
print(lines)
print(type(lines))

```

# Eliminar los saltos de lnea en el archivo importado:
Los tres mtodos que se han visto para leer los archivos importan el salto de lnea. Por lo que puede ser
necesario eliminarlo antes de trabajar con los datos. Esto se puede conseguir de forma sencilla con el
mtodo rstrip de las cadenas de texto de Python. Lo que se puede hacer iterando sobre el vector.

```python
lines1 = [s.rstrip('\n') for s in lines]
lines1
```
