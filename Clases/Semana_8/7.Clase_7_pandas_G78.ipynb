# Union de series mediante la funcion concat

```python
s=pd.Series([1,2,3,4,5],index=['a','b','c','d','e'])
r=pd.Series([10,11,12],index=['f','g','h'])
print(s)
print()
print(r)
```

```python
#La funcion concat hace la concatenacion por defecto en el 
#eje 0 que es el indice de las filas, pero tambien 
# puede unir a partir del eje 1 de las columnas
t=pd.concat([s,r])
print(type(t))
t
```

```python
#Vamos a utilizar la concatenacion en el eje 1 de las columnas, panda alineara los valores por etiquetas identicas
a=pd.Series([1,2,3,4,5],index=['a','b','c','d','e'])
b=pd.Series([10,11,12],index=['a','b','f'])
t=pd.concat([a,b],axis=1,sort=True)
print(type(t))
t
```

```python
#Vamos a utilizar la concatenacion en el eje 1 de las columnas, panda alineara los valores por etiquetas identicas
a=pd.Series([1,2,3,4],index=['a','b','c','d'])
b=pd.Series([10,11,12],index=['a','c','e'])

#Podemos tener etiquetas iguales en las filas de una serie o un 
#dataframe pero se recomiendan que sean unicas y no hayan duplicadas
t=pd.concat([a,b])
print(type(t))
t
```

# Metodo Append

Permite hacer la concatenacion de una forma limitada a  la funcion concat, solamente deja unir en el eje 0

```python
a=pd.Series([1,2,3,4,5],index=['a','b','c','d','e'])
b=pd.Series([10,11,12],index=['f','g','h'])

#Podemos tener etiquetas iguales en las filas de una serie o un 
#dataframe pero se recomiendan que sean unicas y no hayan duplicadas
c=a.append(b)
print(type(c))
c
```

```python
# Atributo Ignore_Index, este atributo hace que cuando 
#se realiza la concatenacion, ignore las etiquetas y por lo tanto 
#los valores de los indices van a estar por defecto desde un numero entero 0
#hasta n-1, donde n es el numero de filas de una serie

d=a.append(b,ignore_index=True)
d
```

# Concatenacion y union de un dataframe

Panda ofrece dos funciones que son concat y merge

* La funcion concact nos permite concatenar un dataframe a partir de un eje determinado
* La funcion merge permite realizar uniones (joins) entre dataframe como se realizan en las base de datos. 
* MEtodo append que esta disponible en la libreria de pandas. Este metodo ofrece una funcionalidad similiar a la funcion concat pero reducida, unicamente al eje 0 o a las  filas



# Funcion concat

Esta va a concatenar dos o mas dataframes a lo largo de un eje, puede ser el eje 0 relacionadas a las filas o el eje 1 relacionas a las columnas

```python

import numpy as np
import pandas as pd
df1=pd.DataFrame(np.arange(9).reshape([3,3]),
                index=['a','b','c'],
                columns=['A','B','C'])

df1
```

```python
import numpy as np

df2=pd.DataFrame(np.arange(12).reshape([4,3]),
                index=['a','b','c','d'],
                columns=['B','C','D'])

df2
```

```python
#Estoy uniendo el dataframe df1 y df2 con la funcion concat en el eje por defecto 0, es decir las filas
print(df1)
print()
print(df2)
pd.merge([df1,df2])
```

```python
#Estoy uniendo el dataframe df1 y df2 con la funcion concat en el eje por defecto 0, es decir las filas
print(df1)
print()
print(df2)
pd.concat([df1,df2],sort=True)
```

```python
#Estoy uniendo el dataframe df1 y df2 con la funcion concat en el eje por defecto 0, es decir las filas
print(df1)
print()
print(df2)
#Vamos a concatenar a partir de las columnas
pd.concat([df1,df2],axis=1,sort=True)
```

```python
# Estos ejemplos anteriores, por defecto son de tipo outer, no estamos uniendo las etiquetas sino que creamos
#unas nuevas filas o columnas dependiendo el eje que se estre trabajando con las etiquetas correspondiente
```

```python
#UTilizar el atributo join que busca una etiqueta comun entre ambos dataframe
#Me muestra un nuevo dataframe a partir unicamente de los datos comunes entre el dataframe df1 y df2
pd.concat([df1,df2],join='inner')
```

```python
#UTilizar el atributo join que busca una etiqueta comun entre ambos dataframe
#Me muestra un nuevo dataframe a partir unicamente de los datos comunes entre el dataframe df1 y df2
pd.concat([df1,df2],axis=1,join='inner')
```

```python
#UTilizar el atributo join que busca una etiqueta comun entre ambos dataframe
#Me muestra un nuevo dataframe a partir unicamente de los datos comunes entre el dataframe df1 y df2
pd.concat([df1,df2],axis=1,join='outer')
```

```python
#UTilizar el atributo join que busca una etiqueta comun entre ambos dataframe
#Me muestra un nuevo dataframe a partir unicamente de los datos comunes entre el dataframe df1 y df2
pd.concat([df1,df2],axis=1,join='inner',ignore_index=True)
```

```python
#Calcular la ganancia de notas mediante un algoritmo que multiplique el numero de 
#notas por un numero entre 1 y 5 y crear una nueva columna con ese valor
import random
def Notas(nota):
    ganancia=nota*random.randint(1,5)
    return ganancia

#Funcion apply, por cada valor en la columna notas le voy a aplicar la funcion que
#se creo llamada Notas y la almacena en la columna que se crea nueva 
#llamada ganancia
df=df.fillna({'notas':1}) # Metodo fillna hace que si en la columna notas encuentra un valor NaN, lo cambia por un valor de 0
df['ganancia']=df['notas'].apply(Notas)

df.head(10)
```

```python
#Funcion que crea una nueva columna diferencia, pero que pasa a la funcion los valores
#de las filas notas y ganancia
def diferencia(fila):
    resultado=fila['notas']-fila['ganancia']
    return resultado

#axis=1 siggnifica que la funcion diferencia se va a aplicar a cada fila del dataframe
#Fila contiene el dataframe df, pero se aplica a las columnsa notas y ganancia
df['diferencia']=df.apply(diferencia, axis=1)
df.head()
```

```python
#Escribir archivos de texto, creamos una lista con los contenidos de cada linea

data=['Linea 1','Linea 2','Linea 3','Linea 4','Linea 5']
```

# Escribir linea a linea

```python
fic=open('text_1.txt','w')
#El metodo write no aade el salto de linea, por lo cual
#es necesario aadirlo ('\n')
for line in data:
    fic.write(line)
    fic.write("\n")
fic.close()
```

```python
# Una forma de escribir el archivo lnea a lnea sin que sea necesario incluir el salto de lnea es con la
# funcin print. Para esto es necesario incluir la opcin "file" con la conexin al archivo
fic=open('text_2.txt','w')

for line in data:
    print(line,file=fic)
fic.close()
```

# Escribir el archivo de una vez:
Finalmente, en el caso de que los datos se encuentren en un objeto iterable se puede utilizar el mtodo writelines para volcar este de golpe. Aunque es necesario tener en cuenta que este mtodo no agrega el salto de lnea, por lo que puede ser necesario agregarlo con antelacin.



```python
fic = open("text_3.txt", "w")
fic.writelines("%s\n" % s for s in data)
fic.close()
```

## Leer el archivo de una vez:
El procedimiento para leer los archivos de texto ms sencillo es hacerlo de una vez con el mtodo
readlines. Una vez abierto el archivo solamente se ha de llamar a este mtodo para obtener el contenido.

```python
fic = open('text_1.txt', "r")
lines = fic.readlines()
fic.close()
lines

# En esta ocasin lines es un vector en el que cada elemento es una lnea del archivo. Alternativamente,
# en lugar del mtodo readlines se puede usar la funcin list para leer los datos
```

```python
fic = open('text_1.txt', "r")
lines = list(fic)
fic.close()
print(lines)
print(type(lines))

```

# Leer el archivo lnea a lnea:
En otras ocasiones puede ser necesario leer el archivo lnea a lnea. Esto se puede hacer simplemente
iterando sobre el fichero una vez abierto. En cada iteracin se podr hacer con cada lnea cualquier
operacin que sea necesaria. En el siguiente ejemplo cada una de las lneas se agrega a un vector.

```python
fic = open('text_1.txt', "r")
lines = []

for line in fic:
    lines.append(line)

fic.close()

lines
```

# Eliminar los saltos de lnea en el archivo importado:
Los tres mtodos que se han visto para leer los archivos importan el salto de lnea. Por lo que puede ser
necesario eliminarlo antes de trabajar con los datos. Esto se puede conseguir de forma sencilla con el
mtodo rstrip de las cadenas de texto de Python. Lo que se puede hacer iterando sobre el vector.

```python
lines1 = [s.rstrip('\n') for s in lines]
lines1
```
