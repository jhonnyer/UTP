```python
import matplotlib.pyplot as plt
x1=[2,5,6,14]
y1=[11,22,33,44]

x2=[2,5,6,15]
y2=[4,12,18,45]

plt.plot(x1,y1,color='blue',linewidth=3,label='Linea 1')
plt.plot(x2,y2,color='green',linewidth=3,label='Linea 2')

plt.title('Dos Gráficas juntas')
#Metodos permiten colocar etiquetas en los ejes del grafico
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.legend()
#MEtodo grid permite añadir una cuadricula o rejilla al grafico
plt.grid()
plt.show()
```

# Grafica de barras

```python
plt.bar(x1,y1,color='blue',linewidth=3,label='Barra 1')
plt.bar(x2,y2,color='green',linewidth=3,label='Barra 2')

plt.title('Dos barras juntas')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.legend()
#Vamos a añadir la cuadricula a la grafica o la rejilla
plt.grid()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
```

# Groupby Series

```python
import pandas as pd
df=pd.DataFrame({
    'Nombre':['John','Maria','Pedro','Jenifer','Bob','Lisa','Jose'],
    'Edad':[23,78,22,19,45,33,20],
    'Ciudad':['Bogota','Medellin','Bogota','Medellin','Bogota','Armenia','Armenia'],
    'Num_niños':[2,0,0,3,2,1,5],
    'Num_mascotas':[5,1,0,5,2,2,5]
})

df
```

```python
df.groupby('Ciudad').nunique()
```

```python
df.groupby('Ciudad')['Nombre'].nunique()
```

```python
df.groupby('Ciudad')['Nombre'].nunique().plot(kind='bar')
plt.show()
```

```python
df.groupby('Ciudad').agg({'Edad':'count'}).plot()
```

```python
df_filtrado=df.groupby('Ciudad').agg({'Edad':'count'})
df_filtrado
```

```python
# Grafica por edades, size muestra un grafico con edades por ciudad de un solo color
df.groupby(['Ciudad','Edad']).size().plot(kind='bar')
```

```python
# Grafica por edades, size muestra un grafico con edades por ciudad de un solo color
#unstack que permite mostrar por cada ciudad un grafico con la de diferentes colores
df.groupby(['Ciudad','Edad']).size().unstack().plot(kind='bar')
```

```python
#rwidth=0.8 la sepracion entre cada barra el histograma
# BINs HAce referencia a las agrupaciones que va a tener el histograma
print(df)
df[['Edad']].plot(kind='hist',bins=[0,20,40,60,80,100],rwidth=0.8)
plt.show()
```

# Dataframe con campo de tipo fecha

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.DataFrame({
    'name':['john','lisa','peter','carl','linda','betty'],
    'date_of_birth':['01/21/1988','03/10/1977','07/25/1999','01/22/1977','09/30/1968','09/15/1970']
})

df
```

```python
df['date_of_birth']=pd.to_datetime(df['date_of_birth'],infer_datetime_format=True)
df
```

```python
#Metodo map me permite obtener una serie a partir de los datos de una fecha de datetime, en este caso los valores 
#Van a ser los meses 
print(df['date_of_birth'].map(lambda d:d.month))
```

```python
df['date_of_birth'].map(lambda d:d.month).plot(kind='hist')
```

```python
import matplotlib.pyplot as plt
from matplotlib.dates import date2num
import datetime

x=[
    datetime.datetime(2011,1,4,0,0),
    datetime.datetime(2011,1,5,0,0),
    datetime.datetime(2011,1,6,0,0)
]

print(x)

x=date2num(x) #Verificar funcion date2num
print()
print(x)
y=[4,9,2]
z=[1,2,3]
k=[11,12,13]
ax=plt.subplot(111)
print(ax)
ax.bar(x-0.2,y,width=0.2,color='b',align='center')
ax.bar(x,z,width=0.2,color='g',align='center')
ax.bar(x+0.2,k,width=0.2,color='r',align='center')
ax.xaxis_date()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
#importar datos para crear dataframe
liquido_url='https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'
liquido_column_headers=['Alcohol','Acido Malico','Cenizas','Alcalinidad de las cenizas',
                        'Magnesio','Fenoles totales,Flavanoides','Fenoles no flavonoides','Proantocianinas',
                       'Intensidad de color','Matiz','0D280/0D315 de sustancia diluida','Prolina']
liquido_df=pd.read_csv(liquido_url,names=liquido_column_headers)
liquido_df.head(3)
```

```python
#Generar los ejes de la grafica de datos con la que vamos a trabajar
fig,ax1=plt.subplots()
print(fig) #Los ejes o tamaños de la grafica
print(ax1) #contiene la grafica donde se va a trabajar
fig.set_size_inches(13,10)

#Sequencia de color
c=liquido_df['Intensidad de color']

#Label o etiquetas
ax1.set_xlabel('Alcohol')
ax1.set_ylabel('Intensidad de color')
ax1.set_title('Relación entre la intensidad de color y la cantidad de alcohol en el liquido')

#Graficar la informacion
plt.scatter(liquido_df['Alcohol'],liquido_df['Intensidad de color'],c=c,cmap='RdPu')

cbar=plt.colorbar()
# https://matplotlib.org/stable/gallery/color/colormap_reference.html
#Añadimos un nombre a la barra
cbar.set_label('Intensidad de color')
```
